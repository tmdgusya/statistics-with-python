```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
import numpy as np
import pandas as pd

numeric_data = pd.DataFrame({
    'length' : [0.5, 1.1, 1.2, 1.3, 1.4, 1.5, 2.1, 2.2, 3.1, 4.6],
})

sns.histplot(
    x='length',
    data=numeric_data,
    color='gray',
    bins=np.arange(0, 6, 1)
)
```

```python
sns.histplot(
    x='length',
    data=numeric_data,
    color='gray',
    bins=np.arange(0, 6, 1),
    stat='density',
)
```

```python
np.histogram(
    numeric_data.length,
    bins=np.array([0,1,2,5]),
    density=True,
)
```

```python
sns.histplot(
    x='length',
    data=numeric_data,
    color='gray',
    bins=np.array([0, 1, 2, 5]),
    stat='density',
)
```

```python
sns.kdeplot(
    numeric_data.length,
    fill=True,
    color='gray',
)
```

```python
sns.kdeplot(numeric_data.length, color='black', label='default')
sns.kdeplot(numeric_data.length, color='black', bw_adjust=0.4, linestyle='dashed', label='bw_adjust=0.4')
sns.kdeplot(numeric_data.length, color='black', bw_adjust=2, linestyle='dotted', label='bw_adjust=2')

plt.legend() # show label on the top of right
```
