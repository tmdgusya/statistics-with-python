```python
from scipy import stats
from matplotlib import pyplot as plt

from matplotlib import rcParams

x = np.arange(start=0, stop=8.1, step=0.1)

density = stats.norm.pdf(x=x, loc=4, scale=0.8)
```

```python
density_df = pd.DataFrame({
    'x': x,
    'density': density
})

density_df.head(3)
```
